```python
def selection(fitness_list):
    #Rank order selection
    selection={}
    rank=[sorted(fitness_list,reverse=True).index(x)+1 for x in fitness_list]
    fitness_rank=dict(zip(rank,fitness_list))
    for i,j in fitness_rank.items():
        selection[j]=(i/(25*24))*100
    return selection
    
    
    
```
